```python
def getMappedDestination(_tuple:list,element:int):
    if(element > _tuple[SOURCE]+_tuple[RANGE]-1) : return element 
    return _tuple[DESTINATION] + element-_tuple[SOURCE]
    
def findLowerBound(_list:list,element:int):
    start = 0
    end = len(_list) - 1
    answer = 0
    
    if(element<_list[0][SOURCE] or element>_list[-1][SOURCE]+_list[-1][RANGE]-1):
        return element
    
    while start <= end:
        middle = (start + end)//2    #Finding the middle point of the list
        if _list[middle][SOURCE] == element:
            return _list[middle][DESTINATION]
        elif _list[middle][SOURCE] < element:
            answer = middle
            start = middle + 1
        else:
            end = middle - 1
    
    return getMappedDestination(_list[answer],element)
```

```python
def getLocation(element):
    for key in maps:
        # print(key,end="\t:")
        # print(element,end="\t")
        element = findLowerBound(maps[key],element)
        # print(element)
    return element
```
