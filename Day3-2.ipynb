```python
def checkForGear(row:int,col:int)->bool:
    gear = set()
    for r in range(row-1,row+2):
        if(r<0 or r>=len(input)): continue
        for c in range(col-1,col+2):
            if(c<0 or c>=len(input[0])): continue
            if(input[r][c] == "*"):
                gear.add((r,c))
    return gear
```
