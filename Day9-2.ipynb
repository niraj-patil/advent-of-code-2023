```python
solution = 0
for history in histories:
    difference_list = [history]
    level = 1
    while sum(difference_list[-1]) != 0:
        difference_list.append([])
        for i in range(len(difference_list[level-1])-1):
            difference_list[level].append(difference_list[level-1][i+1] - difference_list[level-1][i])
        level += 1
        
    next_value = 0  
    for level in difference_list:
        next_value += level[-1]
    
    solution += next_value

solution
```
